```python
import preprocesing
import speakerFeature
import wordFeature
import pandas as pd
```

```python
def create_feature_dataset(audio_files, preprocessor, speaker_extractor, word_extractor):
    all_speaker_features = []
    all_word_features = []
    
    for file_info in audio_files:
        # Preprocesar audio
        audio = preprocessor.process_audio(file_info['ruta_archivo'])
        # Extraer características
        speaker_features = speaker_extractor.process_audio(audio)
        word_features = word_extractor.process_audio(audio)
        
        all_speaker_features.append(speaker_features)
        all_word_features.append(word_features)
            
    return pd.DataFrame(all_speaker_features), pd.DataFrame(all_word_features)

# Crear instancias
preprocessor = preprocesing.AudioPreprocessor()
speaker_extractor = speakerFeature.SpeakerFeatureExtractor()
word_extractor = wordFeature.WordFeatureExtractor()

df = pd.read_csv('dataset.csv')

# Obtener la primera fila del DataFrame
first_row_df = df.head(1)

speaker_df, word_df = create_feature_dataset(first_row_df.to_dict('records'), preprocessor, speaker_extractor, word_extractor)
```
