```python
from preprocesing import AudioPreprocessor
from speakerFeature import SpeakerFeatureExtractor
from wordFeature import WordFeatureExtractor
import pandas as pd
from IPython.display import Audio
```

```python
def create_feature_dataset(audio_files, preprocessor, speaker_extractor, word_extractor):
    all_speaker_features = []
    all_word_features = []
    
    for file_info in audio_files:
        # Preprocesar audio
        audio = preprocessor.process_audio(file_info['ruta_archivo'])
        
        # Extraer caractersticas
        speaker_features = speaker_extractor.process_audio(audio)
        word_features = word_extractor.process_audio(audio)
        
        all_speaker_features.append(speaker_features)
        all_word_features.append(word_features)
            
    return pd.DataFrame(all_speaker_features), pd.DataFrame(all_word_features),audio

# Crear instancias
preprocessor = AudioPreprocessor()
speaker_extractor = SpeakerFeatureExtractor()
word_extractor = WordFeatureExtractor()

df = pd.read_csv('csv/dataset.csv')

# Obtener la primera fila del DataFrame
first_row_df = df.iloc[11:12]

speaker_df, word_df, audio = create_feature_dataset(first_row_df.to_dict('records'), preprocessor, speaker_extractor, word_extractor)

Audio(audio, rate=preprocessor.sample_rate)
```
